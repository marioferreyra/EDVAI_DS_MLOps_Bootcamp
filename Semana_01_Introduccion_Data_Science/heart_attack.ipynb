## 3) Acondicionar dataset 🔧

Eliminamos las columnas `AgeCategory`, `GenHealth`, `Race`, `Diabetic`.
Esto lo hacemos con la finalidad de ver cómo influirán los demás parámetros en el análisis.

La columna `Race` es eliminada porque la consideramos un dato polemico y puede llegar a ser propenso a un sesgo/bias.

```python
df_data_filter = df_data.drop(['AgeCategory', 'GenHealth', 'Race', 'Diabetic'], axis=1)
```

```python
df_data_filter
```

**Valores categóricos a numéricos**

Las siguienes columnas:
- `Smoking`
- `AlcoholDrinking`
- `Stroke`
- `DiffWalking`
- `PhysicalActivity`
- `Asthma`
- `KidneyDisease`
- `SkinCancer`

Tiene solo dos posibles valores de tipo string (Yes/No). Por lo que vamos a covertirlas al tipo `int` (0/1).

```python
class_map = {'No': 0, 'Yes': 1}
boolean_columns = ['Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'PhysicalActivity', 'Asthma', 'KidneyDisease', 'SkinCancer']

for boolean_column in boolean_columns:
    df_data_filter[boolean_column] = df_data_filter[boolean_column].map(class_map)
```

Para la columna `Sex` tenemos dos posibles valores de tipo string (Male/Female). Por lo que tambien vamos a covertirlas al tipo `int` (0/1).

```python
class_gender_map = {'Male': 0, 'Female': 1}
df_data_filter['Sex'] = df_data_filter['Sex'].map(class_gender_map)
```

La columna `HeartDisease` es nuestro target.
Esta columna tiene dos posibles valores `Yes` (la persona tuvo un ataque al corazón) y `No` (la persona no tuvo un ataque al corazón). Por lo que tambien vamos a covertirlas al tipo `int` (0/1).

```python
df_data_filter['HeartDisease']
```

```python
class_map = {'No': 0, 'Yes': 1}
df_data_filter['HeartDisease'] = df_data_filter['HeartDisease'].map(class_map)
```

```python
df_data_filter.sample(5)
```

## 4) Separación input/output o X/Y ✂️

```python
x_data = df_data_filter.drop('HeartDisease', axis=1)
```

```python
x_data
```

```python
y_data = df_data_filter['HeartDisease']
```

```python
y_data
```

```python
y_data.value_counts()
```

## 5) Separación training / test 💪

```python
from sklearn.model_selection import train_test_split 
```

Vamos a utilizar el 20% de los datos para el conjunto de test

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

## 6) Creación del modelo: Random Forest 😎

```python
# Importamos el modelo que queremos
from sklearn.ensemble import RandomForestClassifier 
```

```python
# Creamos 100 decision trees (default)
classifier_model = RandomForestClassifier(n_estimators=100)
```

```python
classifier_model
```

```python
%%time

classifier_model.fit(x_train, y_train)
```

## 7) Predecimos la clase ✅

```python
# En training
pred_tr = classifier_model.predict(x_train)
```

```python
# En testing
pred_ts = classifier_model.predict(x_test)
```

### 7.1 Validamos Training
Generamos un dataframe con la realidad vrs. lo predicho

```python
df_val_tr = pd.DataFrame({'y_train': y_train, 'pred_tr': pred_tr})
```

```python
df_val_tr
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from pprint import pprint
```

```python
# ¿cuántos aciertos?
train_metrics = {
    'accuracy': accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True),
    'precision': precision_score(df_val_tr.y_train, df_val_tr.pred_tr),
    'recall': recall_score(df_val_tr.y_train, df_val_tr.pred_tr),
    'f1_score': f1_score(df_val_tr.y_train, df_val_tr.pred_tr),
}

pprint(train_metrics)
```

### 7.2 Validamos Testing
Generamos un dataframe con la realidad vrs. lo predicho

```python
df_val_ts = pd.DataFrame({'y_test': y_test, 'pred_ts': pred_ts})
```

```python
# ¿cuántos aciertos?
test_metrics = {
    'accuracy': accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True),
    'precision': precision_score(df_val_ts.y_test, df_val_ts.pred_ts),
    'recall': recall_score(df_val_ts.y_test, df_val_ts.pred_ts),
    'f1_score': f1_score(df_val_ts.y_test, df_val_ts.pred_ts),
}

pprint(test_metrics)
```

### Variar número de procesadores

```python
# Creamos 100 decision trees (default)
classifier_model_2 = RandomForestClassifier(n_estimators=5, n_jobs=-1)
```

```python
classifier_model_2
```

```python
%%time

classifier_model_2.fit(x_train, y_train)
```

```python
# En testing
pred_ts_2 = classifier_model_2.predict(x_test)
```

```python
df_val_ts_2 = pd.DataFrame({'y_test': y_test, 'pred_ts': pred_ts_2})
```

```python
df_val_ts_2
```

```python
# ¿cuántos aciertos?
test_metrics = {
    'accuracy': accuracy_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts, normalize=True),
    'precision': precision_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts),
    'recall': recall_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts),
    'f1_score': f1_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts),
}

pprint(test_metrics)
```

Podemos apreciar que al utilizar todos los procesadores disponibles dentro de la computadora para un número de estimadores de 500.
El tiempo de entrenamiento se reduce de 13.9 s (13900ms) a 929 ms, lo cual es una reduccion de un poco mas del 93%.

Pero en termino de accuracy no hay mucho cambio.

Tomamos varias metricas del modelo `accuracy`, `precision`, `recall`, `f1_score`.

Esto con el find de tener mas nocion del comportamiento del modelo.

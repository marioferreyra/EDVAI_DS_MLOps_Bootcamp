## Data Preparation

```python
import numpy as np
import pandas as pd

import funpymodeling
```

```python
# Para este caso nos interesa visualizar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
# Cargamos el training dataset

df_data_train = pd.read_csv("data/train.csv", sep=',', index_col=0)
```

```python
print(df_data_train.columns)
print(df_data_train.shape)
```

```python
df_data_train
```

```python
# Cargamos el test dataset

df_data_test = pd.read_csv("data/test.csv", sep=',', index_col=0)
```

```python
print(df_data_test.columns)
print(df_data_test.shape)
```

```python
df_data_test
```

```python
# Joineamos el training dataset y test dataset

df_data = pd.concat([df_data_train, df_data_test], axis=0)
```

```python
print(df_data.columns)
print(df_data.shape)
```

```python
df_data
```

```python
# Seleccionamos algunas columnas del dataset

columns = [
    'Age', 'Class', 'Inflight wifi service',
    'Ease of Online booking', 'Seat comfort',
    'Checkin service', 'satisfaction',
]

df_data = df_data[columns].copy()
```

```python
# Renombramos algunas columnas

df_data.rename(
    columns = {
        'Inflight wifi service': 'Wifi',
        'Ease of Online booking': 'Booking',
        'Seat comfort': 'Seat',
        'Checkin service': 'Checkin',
    }, 
    inplace=True
)
```

```python
# Transformamos la columna "satisfaction"

class_map = {'neutral or dissatisfied': 0, 'satisfied': 1}
df_data['satisfaction'] = df_data['satisfaction'].map(class_map)
```

```python
df_data
```

```python
# Aplicamos Get Dummies sobre el Dataset

df_data_dummies = pd.get_dummies(df_data) 
```

## Separación de X de Y para entrenamiento

```python
data_x = df_data_dummies.drop('satisfaction', axis=1)
data_y = df_data_dummies['satisfaction']
```

```python
# Guardamos las columnas

import pickle

with open('pickle_files/categories_ohe.pkl', 'wb') as f:
    pickle.dump(data_x.columns, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Spliteamos el dataset
x_train, x_test, y_train, y_test = train_test_split(
    data_x, data_y, test_size=0.3,
)
```

```python
x_train
```

```python
y_train
```

## Creacion y entrenamiento del modelo

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
# Creamos 1000 decision trees
rf_model = RandomForestClassifier(n_estimators=1000, random_state=19)
```

```python
%%time

rf_model.fit(x_train, y_train)
```

## Calculamos algunas metricas del modelo entrenado

```python
# En training (por defecto asume 0.5 como punto de corte)
pred_tr = rf_model.predict(x_train)
```

```python
x_train
```

```python
pred_tr
```

```python
# En testing (por defecto asume 0.5 como punto de corte)
pred_ts = rf_model.predict(x_test)
```

```python
from pprint import pprint

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
train_metrics = {
    'accuracy': accuracy_score(y_train, pred_tr, normalize=True),
    'precision': precision_score(y_train, pred_tr),
    'recall': recall_score(y_train, pred_tr),
    'f1_score': f1_score(y_train, pred_tr),
}

pprint(train_metrics)
```

```python
test_metrics = {
    'accuracy': accuracy_score(y_test, pred_ts, normalize=True),
    'precision': precision_score(y_test, pred_ts),
    'recall': recall_score(y_test, pred_ts),
    'f1_score': f1_score(y_test, pred_ts),
}

pprint(test_metrics)
```

# Guardamos el modelo entrenado

```python
# Guardar en el disco
filename = 'model/rf.pkl'
pickle.dump(rf_model, open(filename, 'wb'))
```

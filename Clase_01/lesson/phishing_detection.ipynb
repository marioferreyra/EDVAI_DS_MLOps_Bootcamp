## 3) Acondicionar dataset 🔧

**Eliminar columna url**
<br> No tiene un patrón, no nos aporta al análisis.

````python
filter_data = data.drop('url', axis=1)
````

Si se desea eliminar más columnas, lo que se hace es añadirlas todas dentro de una lista [].
<br> Por ejemplo:
<br> filter_data = data.drop(['url','nb_or','ratio_nullHyperlinks], axis=1)


**Valores categóricos a numéricos**
<br> Si la url es legítima le asignamos un 0, caso contrario un 1.

````python
class_map = {'legitimate':0, 'phishing':1}
filter_data['status'] = filter_data['status'].map(class_map)
````

````python
filter_data.head(5)
````

## 4) Separación input/output o X/Y ✂️

````python
x_data=filter_data.drop('status', axis=1)
````

````python
y_data=filter_data['status']
````

*Nota:* Es común ver el "dropeo" de columnas con slicing/iloc. 
<br>No lo aconsejo porque no somos explicitos con la columna que estamos eliminando. 
<br>Además estaríamos asumiendo que la variable a predecir esta al final, y a veces no ocurre.

Ej:

`x_data=data.iloc[:, :-1]`

`y_data=data.iloc[:, -1]`

## 5) Separación training / test 💪

````python
from sklearn.model_selection import train_test_split 
````

````python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3)

print(x_train.shape, y_train.shape)
````

````python
print(x_test.shape, y_test.shape)
````

## 6) Creación del modelo: Random Forest 😎

````python
# Importamos el modelo que queremos
from sklearn.ensemble import RandomForestClassifier 
````

````python
# Creamos 100 decision trees
rf = RandomForestClassifier()

rf.fit(x_train, y_train)
````

## 7) Predecimos la clase ✅

````python
# En training
pred_tr=rf.predict(x_train)

# En testing
pred_ts=rf.predict(x_test)
````

### 7.1 Validamos Training
Generamos un dataframe con la realidad vrs. lo predicho

````python
df_val_tr=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr})

# ¿cuántos aciertos?
sum(df_val_tr.y_train==df_val_tr.pred_tr)/len(df_val_tr)
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
````

### 7.2 Validamos Testing
Generamos un dataframe con la realidad vrs. lo predicho

````python
df_val_ts=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts})
````

````python
# ¿cuántos aciertos?
accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True)
````

## Extra 🎯

### Variar número de estimadores

````python
# Creamos 5 decision trees
rf_1 = RandomForestClassifier(n_estimators=5)

rf_1.fit(x_train, y_train)
````

````python
# En testing
pred_ts_1=rf_1.predict(x_test)
````

````python
df_val_ts_1=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts_1})
````

````python
# ¿cuántos aciertos?
accuracy_score(df_val_ts_1.y_test, df_val_ts_1.pred_ts, normalize=True)
````

### Variar número de procesadores

````python
# Creamos 100 decision trees
rf_2 = RandomForestClassifier(n_jobs=-1)

rf_2.fit(x_train, y_train)
````

````python
# En testing
pred_ts_2=rf_2.predict(x_test)
````

````python
df_val_ts_2=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts_2})
````

````python
# ¿cuántos aciertos?
accuracy_score(df_val_ts_2.y_test, df_val_ts_2.pred_ts, normalize=True)
````

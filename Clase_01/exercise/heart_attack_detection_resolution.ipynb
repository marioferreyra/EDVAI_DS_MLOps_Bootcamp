## 3) Acondicionar dataset 🔧

**Eliminar columnas**
<br> No las consideramos para este análisis.

```python
filter_data = data.drop(['AgeCategory', 'GenHealth', 'Race', 'Diabetic'], axis=1)
```

**Valores categóricos a numéricos**
<br> Si es 'No' le asignamos un 0, caso contrario un 1.

```python
class_map = {'No':0, 'Yes':1}
columns_booleans = ['Smoking','AlcoholDrinking','Stroke','DiffWalking','PhysicalActivity','Asthma','KidneyDisease','SkinCancer']

for name_column in columns_booleans:
    filter_data[name_column] = filter_data[name_column].map(class_map)
```

Si es 'Male' le asignamos un 0, caso contrario un 1.

```python
class_gender_map = {'Male':0, 'Female':1}
filter_data['Sex'] = filter_data['Sex'].map(class_gender_map)
```

```python
filter_data.head(5)
```

## 4) Separación input/output o X/Y ✂️

```python
x_data=filter_data.drop('HeartDisease', axis=1)
```

```python
y_data=filter_data['HeartDisease']
```

*Nota:* Es común ver el "dropeo" de columnas con slicing/iloc. 
<br>No lo aconsejo porque no somos explicitos con la columna que estamos eliminando. 
<br>Además estaríamos asumiendo que la variable a predecir esta al final, y a veces no ocurre.

Ej:

`x_data=data.iloc[:, :-1]`

`y_data=data.iloc[:, -1]`

## 5) Separación training / test 💪

```python
from sklearn.model_selection import train_test_split 
```

20% del dataset para el test

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)

print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

## 6) Creación del modelo: Random Forest 😎

```python
# Importamos el modelo que queremos
from sklearn.ensemble import RandomForestClassifier 
```

```python
# Caso 1
# rf = RandomForestClassifier()

# Caso 2
rf = RandomForestClassifier(n_estimators=500, n_jobs=-1)

rf.fit(x_train, y_train)
```

## 7) Predecimos la clase ✅

```python
# En training
pred_tr=rf.predict(x_train)

# En testing
pred_ts=rf.predict(x_test)
```

### 7.1 Validamos Training
Generamos un dataframe con la realidad vrs. lo predicho

```python
df_val_tr=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr})

# ¿cuántos aciertos?
sum(df_val_tr.y_train==df_val_tr.pred_tr)/len(df_val_tr)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
```

### 7.2 Validamos Testing
Generamos un dataframe con la realidad vrs. lo predicho

```python
df_val_ts=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts})
```

```python
# ¿cuántos aciertos?
accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True)
```

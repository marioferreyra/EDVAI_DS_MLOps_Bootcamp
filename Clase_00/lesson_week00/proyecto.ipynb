```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Cargar data y visualizarla 游

```python
data = pd.read_csv("./layoffs.csv")
```

```python
data.head(5)
```

```python
data.columns
```

## An치lisis de datos 游뱁

```python
data["Fecha comienzo"] = pd.to_datetime(data["Fecha comienzo"], format="%m/%d/%Y")
data["Fecha fin"] = pd.to_datetime(data["Fecha fin"], format="%m/%d/%Y")
```

```python
data["Dias transcurridos"] = (data["Fecha fin"] - data["Fecha comienzo"]).dt.days
```

```python
data.head(5)
```

## Visualizaci칩n 游눶

```python
plt.rcParams["figure.figsize"] = (10,4)

plt.bar(data['Area'], data['Dias transcurridos'])
plt.suptitle('Permanencia en la empresa seg칰n el 치rea')
```

```python
plt.rcParams["figure.figsize"] = (10,4)

plt.bar(data['Manager'], data['Dias transcurridos'])
plt.suptitle('Permanencia en la empresa seg칰n el manager')
```

```python
plt.rcParams["figure.figsize"] = (10,4)

plt.bar(data['Grupo recruitment'], data['Dias transcurridos'])
plt.suptitle('Permanencia en la empresa seg칰n equipo de recruiters que te evaluaron')
```

```python
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
df1 = pd.read_csv(
    "data/Ultimas_Desvinculaciones-Total.csv",
    dtype={
        'employee_full_name': str,
        'area': str,
        'duration_days': int,
        'level': str,
        'recruitment_group': str,
        'recruitment_time': int,
        'avg_rs_initial': float,
        'avg_rs_competition': float,
        'manager_full_name': str,
        'do_course': str,
    },
    parse_dates=[
        'date_start',
        'date_end',
    ],
)
print(df1.shape)
print(df1.columns)
display(df1.dtypes)
display(df1.sample(30))
```

```python
df1['is_avg_rs_competition_better'] = df1['avg_rs_competition'] > df1['avg_rs_initial']
```

```python
df1
```

```python
df1['area'].value_counts(normalize=False), df1['area'].value_counts(normalize=True) * 100
```

```python
df1['area'].value_counts().plot.bar()
```

### La area de **IT** es la que mas desvinculaciones tuvo (8), seguida del area **Ventas** (6) y **Compras** (5)

```python
df1['manager_full_name'].value_counts(normalize=False), df1['manager_full_name'].value_counts(normalize=True) * 100
```

```python
df1['manager_full_name'].value_counts().plot.bar()
```

### Hay 3 managers que concentran mas del 50% de las desvinculaciones:
### * Ricardo Ronaldo (20%)

### * Alejandro Pitorino (16.6%)

```python
df1[
    df1['manager_full_name'].isin([
        'Ricardo Ronaldo',
        'Marina Pitana',
        'Alejandro Pitorino',
    ])
][['manager_full_name', 'area']].value_counts()
```

### Podemos apreciar que estos managers concentran la mayor o la totalidad de desvinculaciones de las areas con mas desvinculadas:
### * Ricardo Ronaldo - Ventas 6/6 (100%)

```python
df1_grouped = df1.groupby(['area']).aggregate(
    {
        'duration_days': 'mean',
    }
)

print(type(df1_grouped))
print(df1_grouped.shape)
display(df1_grouped)
```

```python
df1_grouped = df1_grouped.reset_index()
```

```python
df1_grouped.sort_values(by='duration_days', ascending=True)
```

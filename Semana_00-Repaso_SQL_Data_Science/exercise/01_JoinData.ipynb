```python
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
df1 = pd.read_csv(
    "data/Ultimas_Desvinculaciones-RowData.csv",
    dtype={
        'Id': int,
        'Nombre': str,
        'Apellido': str,
        'Area': str,
        #'Fecha comienzo': str,
        #'Fecha fin': str,
        'Nivel': str,
        'Grupo recruitment': str,
        'Tiempo recruitment': int,
        'Rango Salarial': str,
        'RS competencia': str,
        'Manager': int,
        'Realizo Cursos': str,
    },
    converters={
        'Fecha comienzo': lambda x: datetime.strptime(x, "%m/%d/%Y"),
        'Fecha fin': lambda x: datetime.strptime(x, "%m/%d/%Y"),
    },
)
print(df1.shape)
print(df1.columns)
display(df1.dtypes)
display(df1.sample(30))
```

```python
df1.rename(columns={
    'Id': 'id_employee',
    'Nombre': 'name',
    'Apellido': 'last_name',
    'Area' : 'area',
    'Fecha comienzo': 'date_start',
    'Fecha fin': 'date_end',
    'Nivel': 'level',
    'Grupo recruitment': 'recruitment_group',
    'Tiempo recruitment': 'recruitment_time',
    'Rango Salarial': 'rs_initial',
    'RS competencia': 'rs_competition',
    'Manager': 'id_manager',
    'Realizo Cursos': 'do_course',
}, inplace=True)
```

```python
df1
```

```python
df1['employee_full_name'] = df1['name'] + ' ' + df1['last_name']
df1['duration_days'] = (df1['date_end'] - df1['date_start']).dt.days
```

```python
df1.columns
```

```python
df1 = df1[[
    'employee_full_name', 'area', 'date_start', 'date_end', 'duration_days',
    'level', 'recruitment_group', 'recruitment_time', 'rs_initial',
    'rs_competition', 'id_manager', 'do_course',
]]
```

```python
df1
```

```python
df2 = pd.read_csv(
    "data/Ultimas_Desvinculaciones-Rangos.csv",
    dtype={
        'Rango': str,
        'desde': int,
        'hasta': int,
    }
)
print(df2.shape)
print(df2.columns)
display(df2.dtypes)
display(df2.sample(5))
```

```python
df2['avg_rs'] = (df2['desde'] + df2['hasta']) / 2
```

```python
df2
```

```python
df12 = pd.merge(
    df1, df2[['Rango', 'avg_rs']],
    left_on='rs_initial',
    right_on='Rango',
    how='left'
)
```

```python
df12
```

```python
df12.columns
```

```python
df12.rename(columns={
    'avg_rs': 'avg_rs_initial',
}, inplace=True)

df12 = df12[[
    'employee_full_name', 'area', 'date_start', 'date_end', 'duration_days',
    'level', 'recruitment_group', 'recruitment_time', 'avg_rs_initial',
    'rs_competition', 'id_manager', 'do_course',
]]
```

```python
df12 = pd.merge(
    df12, df2[['Rango', 'avg_rs']],
    left_on='rs_competition',
    right_on='Rango',
    how='left'
)
```

```python
df12.rename(columns={
    'avg_rs': 'avg_rs_competition',
}, inplace=True)

df12 = df12[[
    'employee_full_name', 'area', 'date_start', 'date_end', 'duration_days',
    'level', 'recruitment_group', 'recruitment_time', 'avg_rs_initial',
    'avg_rs_competition', 'id_manager', 'do_course',
]]
```

```python
df12
```

```python
df3 = pd.read_csv(
    "data/Ultimas_Desvinculaciones-Managers.csv",
    dtype={
        'Id': int,
        'Nombre': str,
        'Apellido': str,
    }
)
print(df3.shape)
print(df3.columns)
display(df3.dtypes)
display(df3.sample(10))
```

```python
df123 = pd.merge(
    df12, df3[['Id', 'manager_full_name']],
    left_on='id_manager',
    right_on='Id',
    how='left'
)
```
